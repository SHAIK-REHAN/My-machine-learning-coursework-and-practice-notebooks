```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df.head()
```

```python
# now  im displaying central tendency 
df.describe()
```

```python
df.columns
```

```python
# now  kind of doing  feature engineering 
df.isnull().sum()
```

```python
# Mean, Median, Mode for all numeric features

mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
mode_values = df.mode(numeric_only=True).iloc[0]

print("Mean Values:\n", mean_values, "\n")
print("Median Values:\n", median_values, "\n")
print("Mode Values:\n", mode_values)
```

```python
#Mean, Median, Mode for ONE particular  column
df['Rented Bike Count'].mean()
df['Rented Bike Count'].median()
df['Rented Bike Count'].mode()[0]
```

```python
#1 Distribution of Rented Bike Count
#this shows how bike rentals are spread across the dataset.
plt.figure(figsize=(7,4))
sns.histplot(df['Rented Bike Count'], kde=True)
plt.title("Distribution of Rented Bike Count")
plt.xlabel("Bike Count")
plt.ylabel("Frequency")
plt.show()
# Shows how frequently different rental counts appear
```

```python
#2.Temperature Distribution
#To see how temperature varies in Seoul across the year.
plt.figure(figsize=(7,4))
sns.histplot(df['Temperature(°C)'], kde=True)
plt.title("Temperature Distribution")
plt.show()
# Shows temperature spread in dataset
```

```python
#3 Humidity Distribution
#Humidity trends that may affect bike usage.
plt.figure(figsize=(7,4))
sns.histplot(df['Humidity(%)'], kde=True)
plt.title("Humidity Distribution")
plt.show()
# Visualizes humidity variation
```

```python
#4.Rented Bike Count vs Temperature
#Checking if warmer temperatures increase rentals.
plt.figure(figsize=(7,4))
sns.scatterplot(x=df['Temperature(°C)'], y=df['Rented Bike Count'])
plt.title("Bike Count vs Temperature")
plt.show()
# Shows a positive correlation between temperature and rentals
```

```python
#5 Rented Bike Count vs Humidity
#Helps see whether humidity affects rentals.
plt.figure(figsize=(7,4))
sns.scatterplot(x=df['Humidity(%)'], y=df['Rented Bike Count'])
plt.title("Bike Count vs Humidity")
plt.show()
# High humidity slightly decreases rentals

```

```python
#6 Rented Bike Count vs Wind Speed
#Windy weather might reduce bike usage.
plt.figure(figsize=(7,4))
sns.scatterplot(x=df['Wind speed (m/s)'], y=df['Rented Bike Count'])
plt.title("Bike Count vs Wind Speed")
plt.show()
# Weak negative relationship with wind speed
```

```python
#7️ Average Bike Rentals by Season
#Find which season has the highest bike usage 
plt.figure(figsize=(7,4))
sns.barplot(x=df['Seasons'], y=df['Rented Bike Count'])
plt.title("Average Rentals by Season")
plt.show()
# Summer/Autumn have higher rentals
```

```python
#8️Rentals on Holiday vs No Holiday
plt.figure(figsize=(6,4))
sns.barplot(x=df['Holiday'], y=df['Rented Bike Count'])
plt.title("Holiday vs Non-Holiday Rentals")
plt.show()
# Non-holidays show more rentals
```

```python
# 9️ Rentals on Functioning Day vs Non-Functioning Day

plt.figure(figsize=(6,4))
sns.barplot(x=df['Functioning Day'], y=df['Rented Bike Count'])
plt.title("Functioning Day vs Rentals")
plt.show()
# Rentals drop sharply on non-functioning days
```

```python
#10 Hour vs Rented Bike Count
#Shows daily rental pattern.

plt.figure(figsize=(8,4))
sns.lineplot(x=df['Hour'], y=df['Rented Bike Count'])
plt.title("Hourly Rental Trend")
plt.show()
# Rentals peak in morning & evening (office hours)
```

```python
### 1️⃣1️⃣ Bike Rentals vs Rainfall
#Rain may reduce rentals.

plt.figure(figsize=(7,4))
sns.scatterplot(x=df['Rainfall(mm)'], y=df['Rented Bike Count'])
plt.title("Rainfall vs Bike Count")
plt.show()
# Higher rainfall decreases rentals
```

```python
#1️2️ Snowfall vs Bike Rentals

plt.figure(figsize=(7,4))
sns.scatterplot(x=df['Snowfall (cm)'], y=df['Rented Bike Count'])
plt.title("Snowfall vs Bike Count")
plt.show()
# Snowfall sharply reduces rentals
```

```python
# 13️ Correlation Heatmap
#This heatmap helps identify how strongly numerical features are related to each other. 
#It highlights positive and negative correlations and shows which variables move together.
plt.figure(figsize=(12,6))
corr = df.corr()

sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()

# Shows strongest relations like Temperature ↔ DewPointTemp and Rentals ↔ Temperature
#Red =  positive correlation
#Blue =  negative correlation
```

```python
corr = df.select_dtypes(include='number').corr()
```

```python
# 13️ Correlation Heatmap
# This heatmap helps identify how strongly numerical features are related to each other.

plt.figure(figsize=(12,6))

# Select only numeric columns for correlation
corr = df.select_dtypes(include='number').corr()

sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()

# Shows strongest relations like Temperature ↔ DewPointTemp and Rentals ↔ Temperature
# Red = positive correlation, Blue = negative correlation
```

```python
# 14️ Monthly Rental Trend

# Extract month from the Date column
df['Month'] = df['Date'].dt.month

plt.figure(figsize=(8,4))
sns.lineplot(x=df['Month'], y=df['Rented Bike Count'])
plt.title("Monthly Rental Trend")
plt.xlabel("Month")
plt.ylabel("Bike Rentals")
plt.show()

# Rentals are highest in warmer months and lowest in winter months
```
