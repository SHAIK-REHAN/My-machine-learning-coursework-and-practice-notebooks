```python
data = ["red", "blue", "?", "green", "blue", "", "N/A", "green", "blue"]

invalid_values = ["?", "N/A", "None", ""]
clean_data = []

for item in data:
    if item not in invalid_values:
        clean_data.append(item)

freq = {}

for item in clean_data:
    if item not in freq:
        freq[item] = 1
    else:
        freq[item] += 1

max_category = None
max_count = 0

for category in freq:
    if freq[category] > max_count:
        max_category = category
        max_count = freq[category]

print(f"Most Frequent Category: {max_category} ({max_count} times)")
```

```python
temps = [30, 32, 31, 29, 28, 35, 36]
k = 3

moving_avg = []

for i in range(len(temps) - k + 1):
    window_sum = 0
    for j in range(i, i + k):
        window_sum += temps[j]
    avg = round(window_sum / k, 2)
    moving_avg.append(avg)

print(moving_avg)
```
